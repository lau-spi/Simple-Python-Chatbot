```python
# %load train_chatbot.py
import nltk
nltk.download('punkt')
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
import json
import pickle

import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout
from keras.optimizers import SGD
import random

words=[]
classes = []
documents = []
ignore_words = ['?', '!']
data_file = open('intents.json').read()
intents = json.loads(data_file)


for intent in intents['intents']:
    for pattern in intent['patterns']:

        # take each word and tokenize it
        w = nltk.word_tokenize(pattern)
        words.extend(w)
        # adding documents
        documents.append((w, intent['tag']))

        # adding classes to our class list
        if intent['tag'] not in classes:
            classes.append(intent['tag'])

words = [lemmatizer.lemmatize(w.lower()) for w in words if w not in ignore_words]
words = sorted(list(set(words)))

classes = sorted(list(set(classes)))

print (len(documents), "documents")

print (len(classes), "classes", classes)

print (len(words), "unique lemmatized words", words)


pickle.dump(words,open('words.pkl','wb'))
pickle.dump(classes,open('classes.pkl','wb'))

# initializing training data
training = []
output_empty = [0] * len(classes)
for doc in documents:
    # initializing bag of words
    bag = []
    # list of tokenized words for the pattern
    pattern_words = doc[0]
    # lemmatize each word - create base word, in attempt to represent related words
    pattern_words = [lemmatizer.lemmatize(word.lower()) for word in pattern_words]
    # create our bag of words array with 1, if word match found in current pattern
    for w in words:
        bag.append(1) if w in pattern_words else bag.append(0)

    # output is a '0' for each tag and '1' for current tag (for each pattern)
    output_row = list(output_empty)
    output_row[classes.index(doc[1])] = 1

    training.append([bag, output_row])
# shuffle our features and turn into np.array
random.shuffle(training)
training = np.array(training)
# create train and test lists. X - patterns, Y - intents
train_x = list(training[:,0])
train_y = list(training[:,1])
print("Training data created")


# Create model - 3 layers. First layer 128 neurons, second layer 64 neurons and 3rd output layer contains number of neurons
# equal to number of intents to predict output intent with softmax
model = Sequential()
model.add(Dense(128, input_shape=(len(train_x[0]),), activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(len(train_y[0]), activation='softmax'))

# Compile model. Stochastic gradient descent with Nesterov accelerated gradient gives good results for this model
sgd = SGD(lr=0.01, decay=1e-6, momentum=0.9, nesterov=True)
model.compile(loss='categorical_crossentropy', optimizer=sgd, metrics=['accuracy'])

#fitting and saving the model
hist = model.fit(np.array(train_x), np.array(train_y), epochs=200, batch_size=5, verbose=1)
model.save('chatbot_model_spacy.h5', hist)

print("model created")
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
import pickle
import numpy as np

from keras.models import load_model
model = load_model('chatbot_model.h5')
import json
import random
intents = json.loads(open('intents.json').read())
words = pickle.load(open('words.pkl','rb'))
classes = pickle.load(open('classes.pkl','rb'))


def clean_up_sentence(sentence):
    sentence_words = nltk.word_tokenize(sentence)
    sentence_words = [lemmatizer.lemmatize(word.lower()) for word in sentence_words]
    return sentence_words

# return bag of words array: 0 or 1 for each word in the bag that exists in the sentence

def bow(sentence, words, show_details=True):
    # tokenize the pattern
    sentence_words = clean_up_sentence(sentence)
    # bag of words - matrix of N words, vocabulary matrix
    bag = [0]*len(words)
    for s in sentence_words:
        for i,w in enumerate(words):
            if w == s:
                # assign 1 if current word is in the vocabulary position
                bag[i] = 1
                if show_details:
                    print ("found in bag: %s" % w)
    return(np.array(bag))

def predict_class(sentence, model):
    # filter out predictions below a threshold
    p = bow(sentence, words,show_details=False)
    res = model.predict(np.array([p]))[0]
    ERROR_THRESHOLD = 0.25
    results = [[i,r] for i,r in enumerate(res) if r>ERROR_THRESHOLD]
    # sort by strength of probability
    results.sort(key=lambda x: x[1], reverse=True)
    return_list = []
    for r in results:
        return_list.append({"intent": classes[r[0]], "probability": str(r[1])})
    return return_list

def getResponse(ints, intents_json):
    tag = ints[0]['intent']
    list_of_intents = intents_json['intents']
    for i in list_of_intents:
        if(i['tag']== tag):
            result = random.choice(i['responses'])
            break
    return result

def chatbot_response(msg):
    ints = predict_class(msg, model)
    res = getResponse(ints, intents)
    return res

#Console chat

chat = True
while chat:
    msg = input().strip()
    if msg=="stop": 
        chat = False
        break
    res = chatbot_response(msg)
    print(res+"\n")
```

## Model

```python
classes = []
documents = []
ignore_words = ['?', '!']
data_file = open('inquiries.json').read()
intents = json.loads(data_file)


for intent in intents['intents']:
    for pattern in intent['patterns']:
        
        # adding documents - spacy !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
        doc = nlp(pattern)
        documents.append((doc, intent['tag']))

        # adding classes to our class list
        if intent['tag'] not in classes:
            classes.append(intent['tag'])

classes = sorted(list(set(classes)))
print (len(documents), "documents")
print (len(classes), "classes", classes)

pickle.dump(classes,open('classes.pkl','wb'))

# initializing training data
training = []
output_empty = [0] * len(classes)
for document in documents: #!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
    doc = document[0]
    vec = doc.vector
    # output is a '0' for each tag and '1' for current tag (for each pattern)
    output_row = list(output_empty)
    output_row[classes.index(document[1])] = 1
    training.append([vec, output_row])
    
# shuffle our features and turn into np.array
random.shuffle(training)
training = np.array(training)
# create train and test lists. X - patterns, Y - intents
train_x = list(training[:,0])
train_y = list(training[:,1])
print("Training data created")


# Create model - 3 layers. First layer 128 neurons, second layer 64 neurons and 3rd output layer contains number of neurons
# equal to number of intents to predict output intent with softmax
model = Sequential()
model.add(Dense(128, input_shape=(len(train_x[0]),), activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(len(train_y[0]), activation='softmax'))

# Compile model. Stochastic gradient descent with Nesterov accelerated gradient gives good results for this model
sgd = SGD(lr=0.01, decay=1e-6, momentum=0.9, nesterov=True)
model.compile(loss='categorical_crossentropy', optimizer=sgd, metrics=['accuracy'])

#fitting and saving the model
hist = model.fit(np.array(train_x), np.array(train_y), epochs=500, batch_size=10, verbose=1)
model.save('chatbot_model_spacy.h5', hist)

print("model created")
```

```python
def predict_class_smart(doc, model):
    # filter out predictions below a threshold
    p = doc.vector #!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
    res = model.predict(np.array([p]))[0]
    #print("\nresults? maybe?")
    #for r in res:
    #    print(r)
    #print("\n")
    ERROR_THRESHOLD = 0.1
    results = [[i,r] for i,r in enumerate(res) if r>ERROR_THRESHOLD]
    # sort by strength of probability
    results.sort(key=lambda x: x[1], reverse=True)
    return_list = []
    for r in results:
        return_list.append({"intent": classes[r[0]], "probability": str(r[1])})
        print("Intents and probs")
        print("intent", classes[r[0]], "probability", str(r[1]))
    print("\n")
    return return_list
```

```python
#Match type of inquiry

matcher = Matcher(nlp.vocab)
matcher.add("date", None, [{"ENT_TYPE":"DATE"}])
matcher.add("time", None,
            [{"lower": {"IN": ["start", "begin"]}}],
            [{'LOWER': {"IN":["at", "before"]}},{'ENT_TYPE': 'TIME'}],
            [{'LOWER': {"IN":["at", "before"]}},{'LIKE_NUM': True}])
# when do we have to be there
matcher.add("genre", None, 
            [{"lemma": {"IN": ["genre","comedy", "horror", "romance", "romantic", "drama", "crime", "tragedy", "thriller", "tragic", "scary", "funny"]}}])
# what kind of movie
matcher.add("cast", None,
            [{"LEMMA": {"IN": ["cast", "star"]}}],
            [{"POS":"PROPN", "lower": {"NOT_IN":["alaska","park","bible","taken","islamabad","london","godfather","harry","potter","greenland","subversion","jurassic","berlin"]}}],
            #[{"lemma":"be"},{"IS_ASCII":True,"OP":"*"},{"lower":"in"}],
            [{"lemma":"do"},{"IS_ASCII":True,"OP":"*"},{"lemma":"appear"}])
matcher.add("runtime", None,
            [{"lower":"runtime"}],
            [{"lower":"how"},{"lower":"long"}],
            [{"LIKE_NUM": True},{"lower": {"IN": ["h","hours"]}}])
matcher.add("synopsis", None, 
            [{"lower": {"IN": ["synopsis","plot", "summary"]}}],
            [{"lower":"what"},{"lower":"happens"}],
            [{"lower":"what"},{"lower":"is"},{"IS_ASCII":True,"OP":"*"},{"lower":"about"}])
matcher.add("recommendation", None, 
            [{"lemma": {"IN": ["recommend", "recommendation"]}}],
            [{"lemma": "be", "OP":"?"},{"lemma": {"IN": ["good", "bad"]}}])



#Match movie

movie_matcher = Matcher(nlp.vocab)
movie_matcher.add("taken", None, [{"lower": "taken"}])
movie_matcher.add("bible", None, [{"lower": "bible"}])
movie_matcher.add("blumen", None, [{"lower": {"IN": ["blumen", "islamabad"]}}])
movie_matcher.add("dishes", None, [{"lower": {"IN": ["dishes", "london"]}}])
movie_matcher.add("godfather", None, [{"lower": "godfather"}])
movie_matcher.add("bread", None, [{"lower": {"IN": ["baking", "bread"]}}])
movie_matcher.add("hp", None, [{"lower": {"IN": ["harry", "potter", "dwarf"]}}])
movie_matcher.add("dictator", None, [{"lower": "dictator"}])
movie_matcher.add("shopping", None, [{"lower": {"IN": ["shopping", "greenland"]}}])
movie_matcher.add("jurassic", None, [{"lower": {"IN": ["jurassic", "subversion"]}}])
movie_matcher.add("birth", None, [{"lower": {"IN": ["birthing", "birth", "berlin"]}}])
```

```python
#Responses

responses = {}
responses["date"] = " is showing on "
responses["time"] = " starts at "
responses["genre"] = " is a "
responses["cast"] = "'s cast includes "
responses["runtime"] = "'s runtime is "
responses["synopsis"] = "'s plot is "
responses["recommendation"] = " is a movie I would "



#Read Movie list

movies = pd.read_csv("movies.tsv",sep="\t")
movies.set_index("movie", inplace=True)
```

```python
CURRENT_MOVIE = "NONE"

def predict_inquiry_class(doc):
    matches = matcher(doc)
    if matches:
        for match_id, start, end in matches:
            string_id = nlp.vocab.strings[match_id]  # Look up string ID
            span = doc[start:end]
            print(string_id,span)
        return nlp.vocab.strings[matches[0][0]]
    return False

def predict_movie(doc):
    matches = movie_matcher(doc)
    if matches:
        for match_id, start, end in matches:
            string_id = nlp.vocab.strings[match_id]  # Look up string ID
            span = doc[start:end]
            #print(string_id,span)
        return nlp.vocab.strings[matches[0][0]]
    return False

def get_response(inq, doc):
    global CURRENT_MOVIE
    res = "sorry, default"
    print(doc)
    
    if not CURRENT_MOVIE == "NONE":
        answer = str(movies.loc[CURRENT_MOVIE][inq])
        print(answer)
        noun=inq
        for token in doc:
            if token.pos_ == "NOUN":
                sim = nlp(inq)[0].similarity(token)
                print(token.text, token.pos_, sim)
                if sim>0.33: noun = token.text
        res = "The "+noun+" is "+answer+"."
        
    
    return res

# parse message and respond
def chatbot_response(doc):
    #inq = predict_inquiry_class(msg)
    inqs = predict_class_smart(doc, model)
    inq = inqs[0]["intent"]
    #print(inq)
    
    global CURRENT_MOVIE
    movie = predict_movie(doc)
    if movie:
        CURRENT_MOVIE = movie
    print("you want to know ",inq," about ",CURRENT_MOVIE)
    res = get_response(inq, doc)
    return res

#Console chat

chat = True
while chat:
    msg = input().strip()
    if msg=="stop": 
        chat = False
        break
    doc = nlp(msg)
    res = chatbot_response(doc)
    print(res+"\n")
```
