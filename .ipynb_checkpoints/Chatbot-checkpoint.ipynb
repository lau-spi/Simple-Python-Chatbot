## Import

```python
import nltk
nltk.download('punkt')
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
import spacy
from spacy import displacy
nlp = spacy.load("en_core_web_lg")
import json
import pickle

import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout
from keras.optimizers import SGD
import random

print("done")
```

## Model

```python
classes = []
documents = []
ignore_words = ['?', '!']
data_file = open('inquiries.json').read()
intents = json.loads(data_file)


for intent in intents['intents']:
    for pattern in intent['patterns']:
        
        # adding documents - spacy !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
        doc = nlp(pattern)
        documents.append((doc, intent['tag']))

        # adding classes to our class list
        if intent['tag'] not in classes:
            classes.append(intent['tag'])

classes = sorted(list(set(classes)))
print (len(documents), "documents")
print (len(classes), "classes", classes)

pickle.dump(classes,open('classes.pkl','wb'))

# initializing training data
training = []
output_empty = [0] * len(classes)
for document in documents: #!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
    doc = document[0]
    vec = doc.vector
    # output is a '0' for each tag and '1' for current tag (for each pattern)
    output_row = list(output_empty)
    output_row[classes.index(document[1])] = 1
    training.append([vec, output_row])
    
# shuffle our features and turn into np.array
random.shuffle(training)
training = np.array(training)
# create train and test lists. X - patterns, Y - intents
train_x = list(training[:,0])
train_y = list(training[:,1])
print("Training data created")


# Create model - 3 layers. First layer 128 neurons, second layer 64 neurons and 3rd output layer contains number of neurons
# equal to number of intents to predict output intent with softmax
model = Sequential()
model.add(Dense(128, input_shape=(len(train_x[0]),), activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(len(train_y[0]), activation='softmax'))

# Compile model. Stochastic gradient descent with Nesterov accelerated gradient gives good results for this model
sgd = SGD(lr=0.01, decay=1e-6, momentum=0.9, nesterov=True)
model.compile(loss='categorical_crossentropy', optimizer=sgd, metrics=['accuracy'])

#fitting and saving the model
hist = model.fit(np.array(train_x), np.array(train_y), epochs=500, batch_size=10, verbose=1)
model.save('chatbot_model_spacy.h5', hist)

print("model created")
```

### Problems to catch:

Using spacy and pyinflect, there are a couple of problems (high similarity of dissimilar words, grammar mistakes when recognizing word type or inflecting, and so on).
These need to be accounted for special.


**Disallowed Words**

Not actually verbs:
Not actually nouns:
Not actually adjectives/adverbs: how, when [what are those, btw?]

Problematic similarity: movie


**Wrong recognition**

- "star" (verb, 3rd p s) sometimes recognized as noun
- starttime is a synonym to duration, but not recognized as start time?


**Wrong inflection**

- to star --> it stares

```python
#Read Movie list

movies = pd.read_csv("movies_2.tsv",sep="\t")
movies.set_index("movie", inplace=True)
movies.head(20)
```

```python
bad_words = ["how", "when", "movie"]
for title in movies["title"]:
    doc = nlp(title)
    for token in doc:
        if token.pos_ in ["PROPN", "NOUN", "VERB", "ADJ", "ADV"]:
            bad_words.append(token.lemma_)
print(bad_words)
```

```python
CURRENT_MOVIE = "NONE"

def predict_class_smart(doc, model):
    # filter out predictions below a threshold
    p = doc.vector #!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
    res = model.predict(np.array([p]))[0]
    ERROR_THRESHOLD = 0.1
    results = [[i,r] for i,r in enumerate(res) if r>ERROR_THRESHOLD]
    # sort by strength of probability
    results.sort(key=lambda x: x[1], reverse=True)
    return_list = []
    for r in results:
        return_list.append({"intent": classes[r[0]], "probability": str(r[1])})
        print("Intents and probs")
        print("intent", classes[r[0]], "probability", str(r[1]))
    return return_list

def predict_inquiry_class(doc):
    matches = matcher(doc)
    if matches:
        for match_id, start, end in matches:
            string_id = nlp.vocab.strings[match_id]  # Look up string ID
            span = doc[start:end]
            print(string_id,span)
        return nlp.vocab.strings[matches[0][0]]
    return False

def predict_movie(doc):
    matches = movie_matcher(doc)
    if matches:
        for match_id, start, end in matches:
            string_id = nlp.vocab.strings[match_id]  # Look up string ID
            span = doc[start:end]
            #print(string_id,span)
        return nlp.vocab.strings[matches[0][0]]
    return False

def get_question_type(doc):
    
    verb = False
    verbs = []
    noun = False
    nouns = []
    adj = False
    adjs = []
        
    if doc[0].dep_ == "aux" and doc[1].text == "you" and doc[2].pos_ == "VERB":
        if doc[3].text == "me": doc = doc[4:]
        elif doc[3].tag_ == "RP":
            if doc[4].text == "me": doc = doc[5:]
            else: doc = doc[4:]
        else: doc = doc[3:]
    
    for token in doc: 
        if not token.lemma_ in bad_words: 
            if token.pos_ in ["VERB"]: #", "AUX"]:
                verb = True
                verbs.append(token)
            if token.pos_ in ["ADJ", "ADV"] and not token.lemma_ in bad_words:
                adj = True
                adjs.append(token)
            if token.pos_ in ["NOUN", "PROPN"]:
                noun = True
                nouns.append(token)
                
    result = {}
    result["verb"] = [verb, verbs]
    result["noun"] = [noun, nouns]
    result["adj"] = [adj, adjs]
    return (result)       

def make_verb_sent(qtype, token_inq, answer):
    verbs = qtype["verb"][1]
    verb = random.choice(verbs)
    main_verbs = [verb for verb in verbs if not (verb.pos_ == "AUX" or verb.dep_ == "aux" or verb.tag_ == "MD")]
    if main_verbs: verb = random.choice(main_verbs)
    v_infl = verb._.inflect("VBZ")
    res = "It "+v_infl+" "
    if token_inq.text in ["starttime", "endtime"]: res += "at "+answer+"."
    else: res += answer+"."
    return res

def make_noun_sent(qtype, token_inq, answer):
    nouns = qtype["noun"][1]
    print (nouns)
    noun = [token_inq, 0.3]
    for noun_option in nouns:
        sim = token_inq.similarity(noun_option)
        print(noun_option.text, sim)
        if sim>noun[1]: noun = [noun_option, sim]
    res = "The "+noun[0].text
    if noun[0].tag_ in ["NN","NNP"]: res += " is"
    else: res += " are"
    if token_inq.text == "synopsis": res += ":"
    res += " "+answer+"."
    return res
            
def get_response(inq, doc):
    global CURRENT_MOVIE
    res = "sorry, default"
    
    if inq=="greeting":
        res = "Hi there, how can I help?"
    elif inq=="movies at Cinemaxx":
        res = "The movies showing at Cinemaxx are:"#
        for title in movies["title"]:
            res += "\n- "+title
    elif inq=="goodbye":
        res = "Have a nice day!"
    elif inq=="thanks":
        res = "Happy to help!"
        
    elif not CURRENT_MOVIE == "NONE":
        
        lookup=inq
        #if inq=="starttime" or inq=="endtime" or inq=="when": lookup="time"
        answer = str(movies.loc[CURRENT_MOVIE][lookup])
            
        print(answer)
        
        token_inq = nlp(inq)[0]
        
        qtype = get_question_type(doc)
        if qtype["verb"][0] == True and qtype["noun"][0] == False and qtype["adj"][0] == False:
            res = make_verb_sent(qtype, token_inq, answer)
        elif qtype["verb"][0] == False and qtype["noun"][0] == True and qtype["adj"][0] == False:
            res = make_noun_sent(qtype, token_inq, answer)
            
        else:
            res = movies.loc[CURRENT_MOVIE]["title"]+responses[inq]+answer
    
    return res

# parse message and respond
def chatbot_response(doc):
    #inq = predict_inquiry_class(msg)
    inqs = predict_class_smart(doc, model)
    inq = inqs[0]["intent"]
    #print(inq)
    
    global CURRENT_MOVIE
    movie = predict_movie(doc)
    if movie:
        CURRENT_MOVIE = movie
    print("you want to know ",inq," about ",CURRENT_MOVIE,"\n\n")
    res = get_response(inq, doc)
    return res
```

```python
# console chat
chat = True
while chat:
    msg = input().strip()
    if msg=="stop": 
        chat = False
        break
    doc = nlp(msg)
    res = chatbot_response(doc)
    print(res+"\n")
```

```python
#Creating GUI with tkinter
import tkinter
from tkinter import *

def send(event=None):
        
    msg = EntryBox.get("1.0",'end-1c').strip()
    EntryBox.delete("0.0",END)

    if msg != '':
        ChatLog.config(state=NORMAL)
        ChatLog.insert(END, "You: " + msg + '\n\n')
        ChatLog.config(foreground="#442265", font=("Verdana", 12 ))
        
        doc = nlp(msg)
        res = chatbot_response(doc)
        ChatLog.insert(END, "Bot: " + res + '\n\n')

        ChatLog.config(state=DISABLED)
        ChatLog.yview(END)
    
base = Tk()
base.title("Chatbot")
base.geometry("400x500")
base.resizable(width=FALSE, height=FALSE)

#Create Chat window
ChatLog = Text(base, bd=0, bg="white", height="8", width="50", font="Arial", wrap="word",)

ChatLog.config(state=DISABLED)

#Bind scrollbar to Chat window
scrollbar = Scrollbar(base, command=ChatLog.yview, cursor="heart")
ChatLog['yscrollcommand'] = scrollbar.set

#Create Button to send message
base.bind('<Return>', send)
SendButton = Button(base, font=("Verdana",12,'bold'), text="Send", width="12", height=5,
                    bd=0, bg="#32de97", activebackground="#3c9d9b",fg='#ffffff',
                    command= send )

#Create the box to enter message
EntryBox = Text(base, bd=0, bg="white",width="29", height="5", font="Arial", wrap="word")
#EntryBox.bind("<Return>", send)


#Place all components on the screen
scrollbar.place(x=376,y=6, height=386)
ChatLog.place(x=6,y=6, height=386, width=370)
EntryBox.place(x=128, y=401, height=90, width=265)
SendButton.place(x=6, y=401, height=90)

base.mainloop()
```
